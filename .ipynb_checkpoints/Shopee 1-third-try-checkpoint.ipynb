```python
# Multimode of List 
# using loop + formula  
#test_list = [2, 1, 2, 2, 1, 3] 

def multimode(test_list):
    res = [] 
    test_list1 = Counter(test_list)  
    temp = test_list1.most_common(1)[0][1]  
    for ele in test_list: 
        if test_list.count(ele) == temp: 
        res.append(ele) 
    res = list(set(res)) 
    return res
    
    # Going far ahead of ourselves
    #return "&".join(str(x) for x in res)
```

```python
conc_list = []
all_buyers_list = []
total_buys_list = []
brushers_list = []

num_rows =brush.shape[0]
for row in range(num_rows):
    shopid_mask = brush["shopid"] == brush["shopid"].iloc[row]
    time_plus_1 = brush["event_time"] <= brush["event_time"].iloc[row] + timedelta(hours=1, minutes=0)
    actual_time = brush["event_time"] >= brush["event_time"].iloc[row]
    test_hour = brush[shopid_mask] [time_plus_1][actual_time]
    
    # Stats Compiling 
    num_unique_buyers = len(test_hour["userid"].unique())
    total_buys_in_hour = test_hour.shape[0]
    all_buyers = test_hour["userid"].tolist()
    
    
    
    if num_unique_buyers != 0:
        concentrate_rate = total_buys_in_hour/ num_unique_buyers

    else:
        concentrate_rate = 0
    #Appends
    conc_list.append(concentrate_rate)
    all_buyers_list.append(all_buyers)
    
    if concentrate_rate >=3:
        brushers_list.append(multimode(all_buyers))
        print("Row Number: ", row)
        print("Concentrate Rate: ", concentrate_rate)
        print("All Buyers: ", all_buyers)
        print("Unique buyers: ", test_hour["userid"].unique().tolist(), "\n")
    else: 
        brushers_list.append("")
    
    # Print Update
    if row%10000==0:
        print("Done with: ", row, "rows")
    
brush["conc_rate"]=conc_list
brush["all_buyers_in_past_hour"]=all_buyers_list
brush["brushers_list"]=brushers_list
    

    
```

```python
## Step 1 Select only relevant frames -- Conc rate more or equal than 3.0

brush_culprits_only = brush[brush["conc_rate"]>=3.0]

```

```python
## Step 2 Select only bad shops

bad_shops = pd.DataFrame(brush[brush["conc_rate"]>=3.0]["shopid"].unique())

bad_shops.columns = ["shopid"]

bad_shops
```

```python
# Step 3 only good shops
bad_shop_list = list(brush[brush["conc_rate"]>=3.0]["shopid"].unique())

good_shops =pd.DataFrame(brush[~brush["shopid"].isin(bad_shop_list)]["shopid"].unique())

good_shops.columns = ["shopid"]

good_shops["userid"]="0"

good_shops
```

```python
# Step 4 create a function to iterate through
```

```python
# using numpy.unique  
import numpy as np 
  
# function to get unique values 
def unique(list1): 
    x = np.array(list1) 
    x = list(np.unique(x)) 
    return [str(i) for i in x]
```

```python
def create_submission(df):
    user_sub_list = []
    for shop_sub in df["shopid"]:
        num_rows = brush_culprits_only.shape[0]
        culprits = []
        for row in range(num_rows):
            if shop_sub == brush_culprits_only.iloc[row]["shopid"]:
                culprits += brush_culprits_only.iloc[row]["brushers_list"]
        
        # Creating final_answers
        if culprits == []:
            answer = 0
        else:
            answer = "&".join(unique(culprits))
        #append to list
        user_sub_list.append(answer)
        
    # Create an extra column for submission    
    df["userid"]=user_sub_list
    return df
```

```python

# Ok. Running this now!! Its gonna be long!
create_submission(bad_shops)
```

```python
submission = pd.concat([bad_shops, good_shops])
```
